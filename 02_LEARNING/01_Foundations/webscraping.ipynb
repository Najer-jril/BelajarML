```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://books.toscrape.com/catalogue/category/books/mystery_3/index.html"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
books = soup.find_all("article", class_="product_pod")
```

```python
data = []

for book in books:
  title = book.h3.a["title"]
  price = book.find("p", class_="price_color").text
  availability = book.find("p", class_="instock availability").text.strip()
  rating = book.p["class"][1]

  data.append({
      "Title": title,
      "Price": price,
      "Availability": availability,
      "Rating": rating
  })

df = pd.DataFrame(data)
df.head()
```

```python
all_books = []

for page in range(1, 51):
  url = f"https://books.toscrape.com/catalogue/page-{page}.html"
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "html.parser")

  books = soup.find_all("article", class_="product_pod")

  for book in books:
    title = book.h3.a["title"]
    price = book.find("p", class_="price_color").text
    availability = book.find("p", class_="instock availability").text.strip()
    rating = book.p["class"][1]

    all_books.append({
        "Title": title,
        "Price": price,
        "Availability": availability,
        "Rating": rating
    })

df = pd.DataFrame(all_books)
df.shape
```

```python
df.to_csv("allbooks.csv", index=False)
```

Pararius

```python
all_flats = []

def safe_extract(element, tag, class_name):
    found = element.find(tag, class_=class_name)
    if found:
        return found.text.strip()
    return None

print("Mulai Scraping...")

for page in range(1, 7):
    url = f"https://www.pararius.com/apartments/utrecht/page-{page}"
    print(f"Scraping halaman: {url}")

    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    flats = soup.find_all("section", class_="listing-search-item")

    for flat in flats:
        title_element = flat.find("h2", class_="listing-search-item__title")
        name = title_element.find("a").text.strip() if title_element else None
        location = safe_extract(flat, "div", "listing-search-item__sub-title")
        price_wrapper = flat.find("div", class_="listing-search-item__price")
        price_element = price_wrapper.find("span", class_="listing-search-item__price-main")
        price = price_element.text.strip() if price_element else None
        area = safe_extract(flat, "li", "illustrated-features__item--surface-area")
        rooms = safe_extract(flat, "li", "illustrated-features__item--number-of-rooms")
        interior_check = flat.find("li", "illustrated-features__item--interior")

        if interior_check:
            interior = interior_check.text.strip()
        else:
            interior = None

        all_flats.append({
            "Name": name,
            "Location": location,
            "Price": price,
            "Area": area,
            "Rooms": rooms,
            "Interior": interior
        })

df = pd.DataFrame(all_flats)

df.to_csv("utrecht_Pararius.csv", index=False)
```
